Matplotlib

```python
#importing matplotlib library
import numpy as np
import matplotlib.pyplot as plt
```

```python
#sine-cos wave
x=np.linspace(-5,5,50)
y=np.sin(x)
z=np.cos(x)

plt.plot(x,y,'g-')
plt.plot(x,z,'r--')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Matplotlib')
plt.show()
```

****

Scatter Plot

```python
x=np.linspace(0,10,30)
y=np.sin(x)

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(x,y,color='b')

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter-plot')
plt.show()
```

****

Bar plot

```python
languages=['English','French','Spanish','Latin','Germen']
people=[100,50,150,40,70]

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.bar(languages,people)

plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.title('Bar-plot')
plt.show()
```

****

Pie Chart

```python
languages=['English','French','Spanish','Latin','Germen']
people=[100,50,150,40,70]

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.pie(people,labels=languages,autopct='%1.1f%%')

plt.title('Pie-Chart')
plt.show()
```
